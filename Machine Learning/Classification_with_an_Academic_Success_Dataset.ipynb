```python
#@title Correlation Matrix
def correlation_plot(df):
  df = pd.get_dummies(df).astype(int)
  corr_mat = df.corr()
  plt.figure(figsize = (20, 20))
  sns.heatmap(corr_mat, annot = True, robust = True, linewidths = 0.2, fmt = '.1f')
```

## **Feature Extraction**

```python
df = pd.get_dummies(train_df).astype(int)
```

```python
corr_mat = df.corr()
```

```python
#@title Columns that influence the target
target_col = corr_mat.loc[['Target_Dropout', 'Target_Enrolled', 'Target_Graduate']]
transpose_target_col = target_col.T
columns_needed = []
for col_name, (value_1, value_2, value_3) in zip(transpose_target_col.index,
                                                 zip(transpose_target_col['Target_Dropout'],
                                                     transpose_target_col['Target_Enrolled'],
                                                     transpose_target_col['Target_Graduate'])):
  if round(value_1, 1) != round(value_2, 1) and round(value_1, 1) != round(value_3, 1) and round(value_3, 1) != round(value_2, 1):
    if col_name not in ['Target_Dropout', 'Target_Enrolled', 'Target_Graduate']:
      columns_needed.append(col_name)
```

```python
columns_needed
```

```python
#@title Box Plot of features
for col in columns_needed:
    plt.figure(figsize = (8, 4))
    sns.barplot(x = col, y = 'Target', data = train_df)
    plt.title(f'Bar plot of {col}')
    plt.tight_layout()
    plt.show()
```

## **Model Creation**

```python
#@title Using encoders
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

enc = LabelEncoder()
sc = MinMaxScaler()

X = sc.fit_transform(train_df[columns_needed])
y = enc.fit_transform(train_df['Target'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 28)
```

```python
train_X.shape, train_y.shape
```

```python
test_X.shape, test_y.shape
```

```python
#@title Creating base model

import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model = xgb.XGBClassifier()

model.fit(train_X, train_y)

pred_y = model.predict(test_X)

accuracy = accuracy_score(test_y, pred_y)
precision = precision_score(test_y, pred_y, average = 'weighted')
recall = recall_score(test_y, pred_y, average = 'weighted')
f1 = f1_score(test_y, pred_y, average = 'weighted')
```

```python
print('Accuracy: ',round(accuracy, 2))
print('Precision: ',round(precision, 2))
print('Recall: ',round(recall, 2))
print('F1: ',round(f1, 2))
```

## **Fine Tuning**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer

params = {
    'n_estimators': np.arange(100, 500, 50),
    'learning_rate': np.arange(0.001, 0.5, 0.001),
    'max_depth': np.arange(1, 10),
    'min_child_weight': np.arange(1, 10, 1),
    'gamma': np.arange(0, 1, 0.1),
    'subsample': np.arange(0.1, 1, 0.1),
    'colsample_bytree': np.arange(0.1, 1, 0.1),
    'reg_alpha': np.arange(0.1, 2, 0.1),
    'reg_lambda': np.arange(0.1, 2, 0.1)
}

xgb_model = xgb.XGBClassifier()

scoring = {
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score, average = 'weighted'),
    'recall': make_scorer(recall_score, average = 'weighted'),
    'f1': make_scorer(f1_score, average = 'weighted')
}

search_cv = RandomizedSearchCV(estimator = xgb_model,
                               param_distributions = params,
                               n_iter = 10,
                               scoring = scoring,
                               refit = 'accuracy',
                               cv = 5,
                               verbose = 1,
                               n_jobs = -1,
                               random_state = 28
                               )

search_cv.fit(train_X, train_y)

print("Best parameters found: ", search_cv.best_params_)
print("Best accuracy score: ", search_cv.cv_results_['mean_test_accuracy'][search_cv.best_index_])
print("Best precision score: ", search_cv.cv_results_['mean_test_precision'][search_cv.best_index_])
print("Best recall score: ", search_cv.cv_results_['mean_test_recall'][search_cv.best_index_])
print("Best F1 score: ", search_cv.cv_results_['mean_test_f1'][search_cv.best_index_])
```

```python
best_model = search_cv.best_estimator_
```

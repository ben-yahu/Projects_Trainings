```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/Colab Files/playground-series-s4e5/train.csv')
```

```python
df_train.head()
```

```python
df_train.isnull().sum()
```

```python
df_train.describe()
```

```python
df_train.info()
```

```python
df_train = df_train.drop('id', axis = 1)
```

```python
df_train.info()
```

```python
corr_m = df_train.corr()
plt.figure(figsize = (15, 10))
sns.heatmap(corr_m, annot = True, fmt = '.1f')
```

```python
scaled_df_train = df_train.copy()
selected_col = df_train.columns.drop('FloodProbability')
for i in selected_col:
  scaled_df_train[i] = scaled_df_train[i]/scaled_df_train[i].max()
```

```python
scaled_df_train
```

```python
X = scaled_df_train.drop('FloodProbability', axis = 1)
y = scaled_df_train['FloodProbability']
```

```python
len(X), len(y)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, train_size = 0.7, random_state = 28)
```

```python
len(train_X), len(train_y)
```

```python
len(test_X), len(test_y)
```

```python
train_X.shape, train_y.shape
```

```python
from tensorflow.random import set_seed as SEED
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.losses import Huber
from tensorflow.keras import Sequential as SQ
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import metrics
```

```python
SEED(28)

model = SQ()
model.add(Dense(20, activation = 'relu'))
model.add(Flatten())
model.add(Dense(100, activation = 'relu'))
model.add(Flatten())
model.add(Dense(100, activation = 'relu'))
model.add(Flatten())
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = Huber(),
              optimizer = SGD(learning_rate = 0.1),
              metrics = [metrics.R2Score()])
```

```python
history = model.fit(train_X, train_y, epochs = 15, validation_data = [test_X, test_y])
```

```python
SEED(28)

model_1 = SQ()
model_1.add(Dense(20, activation = 'relu'))
model_1.add(Flatten())
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Flatten())
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Dense(100, activation = 'relu'))
model_1.add(Flatten())
model_1.add(Dense(1, activation = 'sigmoid'))

model_1.compile(loss = Huber(),
              optimizer = SGD(learning_rate = 0.1),
              metrics = [metrics.R2Score()])

history = model_1.fit(train_X, train_y, epochs = 15, validation_data = [test_X, test_y])
```

```python
pd.DataFrame(history.history).plot()
```
